```python
import os
import json

def path_to_dict(path):
    d = {'name': os.path.basename(path)}
    if os.path.isdir(path):
        d['type'] = "month"
        d['children'] = [path_to_dict(os.path.join(path,x)) for x in os.listdir(path)]
    else:
        d['type'] = "hours"
        d['children'] = [path_to_dict(os.path.join(path,x)) for x in os.listdir(path)]

    return d
data = path_to_dict(path)
print (json.dumps(data))
```

```python
import os
import json

root_list =[]
path = 'assets/'
s3path = "http://dbtzf5ff2y301.cloudfront.net/testimages/"
for root, dirs, files in os.walk(path, topdown=False):
    auxFileFolder = []
    #print(root.split('/'))
    for file in files:
        try:
            fullfile = s3path + root.split('/')[2] + '/' + root.split('/')[3] + '/' + root.split('/')[4] + '/' + root.split('/')[5] + '/' + file
            #print(fullfile)
            auxFileFolder.append(fullfile)
        except:
            print(file)
    print(root, auxFileFolder)
    asset = zip([root],[auxFileFolder])
    root_list.append(asset)
    
print(root_list)
```

```python
assets = []
for a in root_list:
    assets.append(dict(a))
print(assets)
```

```python
# Merge the list in just one
from collections import ChainMap

asset = dict(ChainMap(*assets))
asset
```

```python
# function to get vallue from  key in aseets dict set
def imageInput(month, day, hour):
    route =  'assets/lugar/Montecinos/' + '{}/'.format(month) + '{}/'.format(day) + '{}'.format(hour)
    #print('ROUTE IS', route)
    try:
        imageSet = asset[route]
    except:
        imageSet = []
    return imageSet
```

```python
header = []

for l in assets:
    for k,v in l.items():
        header.append(k)
        
print(header)
```

```python
import numpy as np

# Placeholders for the  keys pairs
lugar = []
months = []
days = []
hours = []


# grave the values into the lists
for i, h in enumerate(header):
    try:
        lugar.append(h.split('/')[2])
        months.append(h.split('/')[3])
        days.append(h.split('/')[4])
        hours.append(h.split('/')[5])
    except Exception as e:
        print(e)
        print(h)
        
# get unique values in this lists
#data = [{"location":lugar, "months":[{"month":month, "days": [{"day": day}]}]}]
location = np.unique(lugar)
months = np.unique(months)
#days = np.unique(days)
```
